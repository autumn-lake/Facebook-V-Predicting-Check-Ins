```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import cross_validation, metrics
from sklearn.grid_search import GridSearchCV

from sklearn.feature_selection import SelectKBest, chi2, f_classif
```

```python
df_train = pd.read_hdf('train_hdfs.h5', 'table')
```

```python
df_train.info()
```

```python
df_train.drop(['row_id','place_id'],axis=1,inplace=True)
```

```python
df_train.info()
```

```python
#modelFit(xgb1, df_train, pred, 'group
target = df_train['group']
df_train.drop(['group'],axis=1,inplace=True)
```

```python
df_train.info()
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train, target, test_size=0.1, random_state=42,stratify=target)
```

```python
rf = RandomForestClassifier(n_estimators=10,max_depth=5,oob_score=True,
                            n_jobs=2,random_state=2,verbose=1,
                            warm_start=True)
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_ 
```

```python
rf.oob_score_ 
```

```python
rf.score(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
print('\nModel Report')
print('Accuracy test : {:.4g}'.format(metrics.accuracy_score(y_test.values, rf.predict(X_test))))
#print('AUC Score (Train): {:f}'.format(metrics.roc_auc_score(dtrain[target], dtrain_predprob)))
```
